```python
def test_range(n):
    if n in range(10,900):
        print("%s is in the range"%str(n))   # %s = n (str)
    else:
        print("The number is outside the given range.")
```

```python
test_range(4)
```

```python
test_range(12)
```

```python
def multiply(numbers):
    total=1                   # multiply = 1 and Addition = 0 is unique number.
    for x in numbers:
        total*=x  # total = total*x
    return total
```

```python
multiply((8,3,5,4,6,2,9,7))
```

```python
multiply((2,3,-4,2,5))
```

```python
def is_even_num(l):
    enum=[]
    for n in l:
        if n%2==0:
            enum.append(n)
    return enum
```

```python
is_even_num([1,2,3,4,5,6,7,8,9])
```

```python
def unique_list(l):
    x=[]
    for a in l:
        if a not in x:
            x.append(a)
    return x
```

```python
unique_list([1,2,3,3,4,5,5,6])
```
