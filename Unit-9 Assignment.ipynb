```python
# Method-2.

def summation(test_tup):
    # Converting into list
    test = list(test_tup)

    # Initializing count
    count = 0

    for i in test:
        count+=i
    return count
```

```python
test_tup = (7,8,9,1,10,7)
summation(test_tup)
```

```python
# Sol-2.

def find_repeated_items(input_tuple):
    input_list = list(input_tuple)      # Convert the tuple to a list.
    repeated_items = []

    for item in input_list:
        if input_list.count(item)>1 and item not in repeated_items:
            repeated_items.append(item)
    return repeated_items
```

```python
input_tuple = (2,4,5,6,3,4,2,6,7,9,4,7,3,8,9)
find_repeated_items(input_tuple)
```
