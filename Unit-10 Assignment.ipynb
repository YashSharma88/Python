```python
# Method-2.

setx = set(['green','blue'])
sety = set(['blue','yellow'])
print("Original set elements : ")
print(setx)
print(sety)
print("\nIntersection of two said sets : ")
setz = setx & sety
print(setz)
```

```python
# Sol-4

def vowel_count(str):
    # Initializing count variable to 0
    count = 0
    # Creating a set of vowels
    vowel = set("aeiouAEIOU")
    for alphabet in str:
        if alphabet in vowel:
            count = 1 + count
    print("No. of vowels : ",count)
```

```python
# Driver code
str = "helloeveryone"

# Function call
vowel_count(str)
```

```python
# Sol-5                 

sn1 = {1,2,3,4,5}
sn2 = {4,5,6,7,8}
print("Original sets are : ")
print(sn1)
print(sn2)
print("Difference of sn1 and sn2 using difference() : ")    # Difference means sn1 m hai but sn2 me nhi h (elements)
print(sn1.difference(sn2))
print("Difference of sn2 and sn1 using difference() : ")    # Difference means sn2 m hai but sn1 me nhi h (elements)
print(sn2.difference(sn1))
print("Difference of sn1 and sn2 using (-) operator : ")
print(sn1-sn2)
print("Difference of sn2 and sn1 using (-) operator : ")
print(sn2-sn1)
```

```python
# Sol-7

x = {1,2,3,4}
y = {4,5,6,7}
z = {8}
print("Original set elements : ")
print(x)
print(y)
print(z)
print("\n confirm two given sets have no element(s) in common : ")
print("\nCompare x and y : ")
print(x.isdisjoint(y))            # isdisjoint mean no items are same in sets.
print("\nCompare x and z :")
print(x.isdisjoint(z))
print("\nCompare y and z :")
print(y.isdisjoint(z))
```
