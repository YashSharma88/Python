```python
# Sol :- 1

def interchange_first_and_last(input_list):
    if len(input_list)>=2:
        input_list[0],input_list[-1] = input_list[-1],input_list[0]

# Input list
my_list = [12,35,9,56,24]
print(my_list)

# Call the function to interchange first and last elements
interchange_first_and_last(my_list)

# print the modified list
print(my_list)
```

```python
# Sol :- 4.

list1 = [10,-21,4,-45,66,-93,1]

pos_count, neg_count = 0,0

for num in list1:
    if num>=0:
        pos_count+=1
    else:
        neg_count+=1

print("Positive numbers in the list : ",pos_count)
print("Negative numbers in the lsit : ",neg_count)
```

```python
list = [1,2,1,2,3,4,5,1,1,2,5,6,7,8,9,9]
unique_list = []
duplicate_list = []

for i in list:
    if i not in unique_list:
        unique_list.append(i)
    elif i not in duplicate_list:
        duplicate_list.append(i)

print(duplicate_list)
```
