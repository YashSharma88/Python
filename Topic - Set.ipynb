```python
# symmetric_difference_update() method :-

# The symmetric_difference_update() method will keep only the elements that are NOT present in both sets.

a = {45,2,34,4,5,35,66,"apple","orange",2.45}
b = {4,5,6,"Cherry",5.67}
a.symmetric_difference_update(b)
print(a)
```

```python
# Symmetric_difference() method :-

# The symmetric_difference() method will return a new set, that contains only the elements that are NOT present in both sets.

a = {45,2,34,4,5,35,66,"apple"}
b = {4,5,6,"Cherry"}
c = a.symmetric_difference(b)
print(c)
```

```python
d = b.symmetric_difference(a)
print(d)
```

```python
# isdisjoint() method :-

# ==> Return True if no item in set x is present in set y.

# Ex-

x = {12,34,53,34,22,54,67,77}
y = {10,23,4,3,5}
z = x.isdisjoint(y)               # True because any value not same in both x and y.
print(z)
```

```python
z = y.isdisjoint(x)
print(z)
```

```python
# issubset() method :-

# ==> Returns whether another set contains this set or not.

# Ex- 

x = {1,2,3,4,5,6}
y = {10,12,13,14,1,2,3,4,5,6}
z = x.issubset(y)                 # True because x all elements are present in y.
print(z)
```

```python
z = y.issubset(x)
print(z)
```

```python
# issuperset() method :-

# Return True if all items set y are present in set x.

# Ex - 

x = {"North","south","python","Cherry","apple",1,2,3}
y = {"apple",1,2,3,"North"}
z = x.issuperset(y)           # True because of y all elements are present is x.
print(z)
```

```python
z = y.issuperset(x)            # Because of x all elements not present in y.
print(z)
```

# Copy() method :-

Que-1. Why you need copy() method?

```python
a = {12,4,5,7,23,34,25}
b = a
a.add(10)
print(a)
print(b)            # Because value are add in both variables.
```

```python
a = {12,4,5,7,23,34,25}
b = a.copy()
a.add(10)
print(a)
print(b)           # Copy() method add value only added in callable variable.
```

# Que-3. Write a python program to find the common items in s list using sets.

```python
ar1 = [1,5,10,20,40,80]
ar2 = [6,7,20,80,100]
ar3 = [3,4,15,20,30,70,80,120]
ar1 = set(ar1)
ar2 = set(ar2)
ar3 = set(ar3)
print(ar1)
print(ar2)
print(ar3)
n = ar1.intersection(ar2)
print(n)
m = n.intersection(ar3)
print(m)
print(list(m))
```
