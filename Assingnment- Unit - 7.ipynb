```python
# Sol-

def check_string(str):
    flag_l = False # l = letter
    flag_n = False # n = number
    for  i in str:
        if i.isalpha():
            flag_l = True
        if i.isdigit():
            flag_n = True
    return flag_l and flag_n
```

```python
check_string("Welcome2ourcountry34")
```

```python
# Sol-

# Method-1.

input_string = "14,625,498.002"

# Initalize an empty result string
result_string = ""

# Iterate through the characters in the input string

for char in input_string:
    if char==",":
        result_string+="."
    elif char==".":
        result_string+=","
    else:
        result_string+=char

# print the swapped string
print(result_string)
```

```python
# Sol - 

sample_str = "A@un2st&i5ve"

def find_digits_char_symbols(sample_str):
    char_count = 0
    digit_count = 0
    symbol_count = 0

    for char in sample_str:
        if char.isalpha():
            char_count+=1
        elif char.isdigit():
            digit_count+=1
        else:
            symbol_count+=1
    print("Total counts of chars, digits and symbols")
    print("Chars = ",char_count, "Digits =",digit_count, "Symbols = ",symbol_count)
```

```python
find_digits_char_symbols(sample_str)
```
