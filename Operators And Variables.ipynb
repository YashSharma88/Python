They are 3 types :-
1. 'and' Operator
2. 'or' Operator

# --> Boolean & Logical Operator :-

1. 'and' Operator :-

--> True and True is True
--> True and False is False
--> False and True is False
--> False and False is False

2. 'Or' Operator :-

--> True or True is True
--> True or False is True
--> False or True is True
--> False or False is False

3. 'Not' Operator :-

--> not(True) is False
--> not(False) is True

==> 'and','or' ---> Binary Operator. (It contain 2 values.)
==> 'not' ---> Unary Operator

```python
print("""For Bitwise not(~) Operator :-\n\nEx-1. a=5 Find ~a=?
Sol'n : Formula ---> (2's Compliment = 1's Compliment + 1) / 2's = 1's + 1
                              8 4 2 1                                                # 0 + 0 = 0
                        5 --> 0 1 0 1                                                  1 + 0 = 1
                       ~5 --> 1 0 1 0 ---> 10 XX This is wrong answer.                 0 + 1 = 1
                                                                                       1 + 1 = 0 ---> Carry = 1.
           Formula ---> 2's = 1's + 1
                              8 4 2 1
                        6 --> 0 1 1 0
                     1's  --> 1 0 0 1
                     2's  -->  +    1    Carry(1) Forward to 0.
                              1 0 1 0 ---> -6 Because -6 and ~5 bits are same here.
                     ~5 = -(5+1)
                     ~5 = -6 Answer.
                     Coding Part --> a=5
                                     print(~5) = -6 Answer.

Ex-2. a=6    Then find ~a=?
Sol'n :                      8 4 2 1
                       6 --> 0 1 1 0
                      ~6 --> 1 0 0 1 ---> -7 Answer.""")
```
