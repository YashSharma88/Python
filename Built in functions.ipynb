# 3. isinstance() Function :-

# The isinstance() function returns True if the specified object is of the specified type, Otherwise False.


```python
# Syntax :-

isinstance(object,type)

# In object only one number will be provide.
# In type multiple values like- int, float, complex, str are provide.
```

```python
print(isinstance(1, int))   # Because 1 is a integer.
```

```python
print(isinstance(2.7,int))  # Because 2.7 is a floating number.
```

```python
print(isinstance(4.9, (int,float)))  # Because 4.9 is a floating number.
```

```python
print(isinstance(3+7j, (int,float)))   # 3+7j is a complex number but here mention only int and float so this is False.
```

```python
print(isinstance("python is very amazing", (int, str, float)))
```

# Power / pow() function :-

```python
# The pow() function returns the value of x to the power of y(xy).
# If the third parameter is present, it returns x to the power of y, modulus z.
```

# pow(x,y,z) - x raise to the power of y and divide by z.

```python
# pow(x,y)
# Ex-1.
pow(2,4)    # Because x^y = 2^4 = 16
```

```python
# Ex-2.
pow(3,6,7)    # Because 3**6 / 7 = 729 / 7 = 1 --> Remainder.
```

```python
# Ex-3.
pow(4,3,8)    # Because 4**3 / 8 = 64/8 = 0 --> Remainder
```

```python
pow(3,6,8)    # Because 3**6 / 8 = 729 / 8 = 1 --> Remainder.
```

```python
pow(3+4j,2)      # Formula -[(a+b)*2 = a*2 + b*2 + 2*a*b], (j)*2 = -1
```

```python
pow(2,4,7)     # Because 2**4 / 7 = 16 /7 = 2 --> Remainder
```

```python
# Que-1. 24 raise to the power 2,3,4,5 and divide it by 7. Find the remainder in each case.

a=pow(24,2,7)
b=pow(24,3,7)
c=pow(24,4,7)
d=pow(24,5,7)
print(a,b,c,d)      # 2 6 4 5 ---> Remainders
```

```python
print(a+b+c+d)
```

```python
# Que-2. 63 raise to the power 45 and divide it by 52. Find the remainder.

a=pow(63,45,52)
print(a)
```

# 6. Sorted function :- 
# The sorted() function returns a sorted list of the specified iteable object.

```python
l=[16,62,73,26,45,75,82]
sorted(l)  # By default assending.
```

```python
sorted(l,reverse=True)   # For desending.
```

```python
# Que-1. Take 5 values from the user and arrange them in assending and desending order.

a=int(input("Enter first value : "))
b=int(input("Enter second value : "))
c=int(input("Enter third value : "))
d=int(input("Enter forth value : "))
e=int(input("Enter fifth value : "))
x = (a,b,c,d,e)
print(x)
print(sorted(x))  # For assending
print(sorted(x,reverse=True)) # For desending
```
