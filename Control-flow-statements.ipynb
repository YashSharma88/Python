# 1. if condition :- 
# The if statement is used to test a specific condition, if the condition is True, a block of code (if-block) will be executed.

# Que-1. Take two numbers from the user and print the greater one.

```python
a=int(input("Enter the first value : "))
b=int(input("Enter the second value : "))

if a>b:
    print("The greater value is a",a)
if b>a:
    print("The greater value is b",b)
```

# Que-2. Enter the value from the user and check the number is Even or Odd.

```python
a=int(input("Enter the value : "))

if a%2==0:
    print("The number entered by the user is even number.")
if a%2!=0:
    print("The number entered by the user is odd number.")
```

# 2. if-elif condition :-

# Que-1. Take a value from the user, and if the value will be greater than 20, print "Magic" and if the value is not greater than 20, print "No Magic".

```python
a=int(input("Enter a number : "))

if a>20:
    print("Magic")
else:
    print("No Magic")
```

# 3. if-elif-else condition :-

# Que-1. Enter two value from the user and if the value will be greater print "Greater", if the value will be not greater print "Not Greater" and if the value are equal print "Equal".

```python
a=int(input("Enter the first value : "))
b=int(input("Enter the second value : "))

if a>b:
    print("The greater value is",a)
elif a==b:
    print("The values are equal")
else:
    print("The value is less.")
```

# Que-2. Take a value from the user and check it is divisible by 2 or 3 or neither 2 or 3.

```python
a=int(input("Enter the value : "))

if a%2==0:
    print("Number is divisible by 2.")
elif a%3==0:
    print("Number is divisible by 3.")
else:
    print("Number is not divisible by 2 or 3.")
```

# Que-3. Write a program and take the number from the user in between 1 to 7, if number is 1, print "Monday", number is 2 print "Tuesday" and so on.

```python
a=int(input("Enter the number : "))

if a==1:
    print("Monday")
elif a==2:
    print("Tuesday")
elif a==3:
    print("Wednesday")
elif a==4:
    print("Thusday")
elif a==5:
    print("Friday")
elif a==6:
    print("Saturday")
elif a==7:
    print("Sunday")
else:
    print("Please enter a number in between 1 to 7.")
```

# 4. Nested-if condition :-

# Que-1. Enter the value from the user and check is it greater than 10 or 20 and not.

```python
a=int(input("Enter the value : "))

if a>10:
    print("Inside the if condition.")
    if a>20:
        print("Inside the nested-if condition.")
    else:
        print("Inside the else part of nested-if condition.")
else:
    print("Outside the condition.")
```

# Que-2. Enter user, number of class attended by student, number of assignment submitted by student, if attendance will be greater than 75 then only student can appear in exam one, if student attendence is greater than 75 and submitted more than 10 assignment than only appear in the exam two and if student attendence is less than 75 than suspended.

```python
a=int(input("Enter the number of classes attended by student."))
b=int(input("Enter the number of assignment submitted by the student."))

if a>75:
    print("Eligible for exam one.")
    if b>10:
        print("Eligible for exam one and two.")
    else:
        print("Eligible for exam one but not eligible for exam two.")
else:
    print("You are suspended.")
```
