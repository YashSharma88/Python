# 1. Concatenation of tuples :-

```python
tuple1 = (12,13,14,15)
tuple2 = (10,15,20,25,30)
print(tuple1+tuple2)
```

# 2. Nesting of tuples :-

```python
tuple1 = (20,13,14,15)
tuple2 = (-1,-2,-3,-4)
tuple3 = (tuple1+tuple2)

print(tuple1)
print(tuple2)
print(tuple3)
```

# 17.Using loops in python :-

```python
# Sol :- 1.

a = (14,6,18,23,-5,-9,-20,26,31,-2,8,-10)

n1 = []
n2 = []
a = list(a)     #  Convert tuple into the list.

for i in range(len(a)):
    if a[i]>=0:
        n1.append(a[i])
print(n1)
print(tuple(n1))

for i in range(len(a)):
    if a[i]<=0:
        n2.append(a[i])
print(n2)
print(tuple(n2))
```

```python
# Sol :- 2.

a = (2,5,3,6,4,8,1)
b = []
print(type(b))

a = list(a)
print(a)
print(type(a))

for i in range(len(a)-1):
    b.append(a[i]*a[i+1])
print(b)
print(tuple(b))
```

```python
# Sol :- 3.

a = [(1,2),(2,3),(3,4)]
b = []
print(type(a))

n1 = list(a[0])
n2 = list(a[1])
n3 = list(a[2])

b.append(n1)
b.append(n2)
b.append(n3)
print(b)
```

```python
a = [(1,2),(2,3),(3,4)]

n1 = a[0]
m1 = n1[0]+n1[1]
print(m1)

n2 = a[1]
m2 = n2[0]+n2[1]
print(m2)

n3 = a[2]
m3 = n3[0]+n3[1]
print(m3)

b = []

b.append(m1)
b.append(m2)
b.append(m3)
print(b)
```
