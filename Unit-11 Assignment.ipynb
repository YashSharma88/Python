```python
# Sol-1.

# Method-1.

dic1 = {1:10, 2:20}
dic2 = {3:30, 4:40}
dic3 = {5:50, 6:60}

result_dict = {}        # Create an empty dictionary

result_dict.update(dic1)
result_dict.update(dic2)
result_dict.update(dic3)
print(result_dict)
```

```python
# Method-2.

dic1 = {1:10, 2:20}
dic2 = {3:30, 4:40}
dic3 = {5:50, 6:60}

result_dict = {}

for d in [dic1,dic2,dic3]:
    result_dict.update(d)
print(result_dict)
```

```python
# Sol-2.

d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

def is_key_present(x):
    if x in d:
        print("Key is present in the dictionary.")
    else:
        print("Key is not present in the dictionary.")
```

```python
is_key_present(5)
is_key_present(8)
```

```python
# Sol-3.

# Input dictionary
my_dict = {'x' : 500, 'y' : 5874, 'z' : 560}

# find the maximum and minimum value in the dictionary
max_value = max(my_dict.values())
min_value = min(my_dict.values())

print(f"Maximum value : {max_value}")
print(f"Minimum value : {min_value}")
```

```python
# Sol-4.

my_dict = {'data1' : 100, 'data2' : -54, 'data3' : 247}
print(sum(my_dict.values()))
```

```python
# Sol-5.

my_dict = {'a' : 1, 'b' : 2, 'c' : 3, 'd' : 4}
print(my_dict)

if 'a' in my_dict:
    del my_dict['a']
print(my_dict)
```

```python
# Sol-6. 

colour_dict = { 'red' : '#FF0000',
               'green' : '#008000',
               'black' : '#000000',
                'white' : '#FFFFFF'}

for key in sorted(colour_dict):
    print("%s : %s" %(key,colour_dict[key]))
```
