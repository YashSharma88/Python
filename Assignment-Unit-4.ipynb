# Que-3. A toy vendor supplies three types of toys.
# Battery based toys
# Key based toys
# Electrical charging based toys

# The vendor gives a discount of 10% on orders for battery -based toys, if the order is for more than Rs. 1000.
# On order of more than Rs. 100 for key-based toys, discount of 5% is given.
# On order for electrical-based toys of value more than Rs. 500, discount of 10% is given.

# Assume that the numric codes 1,2,3 are used for battery-based, key-based and electrical-based toys respectively.
# Write a program that reads the product code and the order amount and print out the net amount that the customer is required to pay after the discount.

```python
# Input product code and order amount

product_code = int(input("Enter the product code (1 for battery-based, 2 for key-based, 3 for electrical-based) : "))
order_amount = float(input("Enter the order amount in Rs. : "))

# Initialize discount and net_amount variables

discount=0
net_amount=order_amount

# Check the product code and apply discounts accordingly

if product_code==1:    # For Battery-based toys
    if order_amount>1000:
        discount=0.10
elif product_code==2:  # For Key-based toys
    if order_amount>100:
        discount=0.05
elif product_code==3:  # For Electrical-based toys
    if order_amount>500:
        discount=0.10

# Calculate the net amount after apply the discount (net amount - order amt*discount)

net_amount-=order_amount*discount

# print the net amount the customer is required to pay

print(f"Net amount after discount Rs {net_amount}")
```

# Que-4. The paschim gujrat vij company ltd. computes the electricity bill based on the followwing matrix.

# Units consumed charges
# 0-100 0.50 per unit
# 101-200 Rs. 50 + Rs. 1 per unit over 100 units
# 201-300 Rs. 150 + Rs. 1.50 per unit over 200 units
# 300 Rs. 300 + Rs. 2 per unit over 300 units

# 1. Ask the user to enter the past meter reading and current meter reading.
# 2. Find the units consumed.
# 3. Compute the bill according to give matrix.

```python
# Step-1. Ask the user to enter past and current meter readings.

past_reading = int(input("Enter the past meter reading : "))
current_reading = int(input("Enter the current meter reading : "))

# Step-2. Calculate units consumed

units_consumed = current_reading-past_reading

# Step-3. Computes the bill according to the given matrix

if units_consumed<=100:
    bill=units_consumed*0.50
elif units_consumed<=200:
    bill=50 + units_consumed*1.00
elif units_consumed<300:
    bill=150 + units_consumed*1.50
else:
    bill=300 + units_consumed*2.00

# Step-4. Print the electricity bill

print("Units Consumed",units_consumed)
print("Electricity bill",bill)
```
