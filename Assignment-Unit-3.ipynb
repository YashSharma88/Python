# Que-1. In an experiment the scientist was measuring the length of a line, the following are the lengths in mm 2.398, 5.639, 7.452, 0.123, 5.754. find the average length and also round off upto 2 digits.

```python
length = 2.398+5.639+7.452+0.123+5.754
average_length = length/5
print("The average length is",average_length)
a=round(average_length,2)
print("The round off average length is",a)
```

# Que-2. Rahul travelled a distance of 12567.7854 km. divide it the cost 45,78,63 rs. Find the average of quotient and reminder and figured out the round off value of the quotient and reminder.

```python
a1 = divmod(12567.7854,45)
a2 = divmod(12567.7854,78)
a3 = divmod(12567.7854,63)
avg_quotient = (a1[0] + a2[0] + a3[0])/3
print("The average quotient is",avg_quotient)
print("The round off value of the average quotient is",round(avg_quotient))
avg_remainder = (a1[1] + a2[1] + a3[1])/3
print("The average remainder",avg_remainder)
print("The remainder off value of the average remainder is",round(avg_remainder))
```

# Que-4. Enter the values of 5 stocks in the form of float from the user and find the average of the 5 stocks and finally round off.

```python
a=float(input("Enter the value of the stock first : "))
b=float(input("Enter the value of the stock second : "))
c=float(input("Enter the value of the stock third : "))
d=float(input("Enter the value of the stock forth : "))
e=float(input("Enter the value of the stock fifth : "))
avg = (a+b+c+d+e)/5
print("The average value of the stock is",avg)
print("The round off value of the average stock is",round(avg))
```
