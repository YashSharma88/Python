# Que-1. Take the number from the user and print all the numbers in between 1 to that number.

```python
a=int(input("Enter the number : "))

i=1
while i<a:
    i=i+1
    print(i)
print("Yes, it is done.")
```

```python
# Type-1.

x=int(input("Enter the number of rows : "))

i=1
while i<=x:
    j=1
    while j<=i:
        print("*",end=" ")
        j=j+1
    print()
    i=i+1
```

# Que-1. Write a program to print the square of number till number 10.

```python
L = []
for i in range(10):
    print(i+1)
    L.append(i**2)       # append function used to add or fill empty list, Tuple or String.
print(L)

# Type-2

L = []
for i in range(0,10,2):
    print(i+1)
    L.append(i**2)
print(L)
```

# Que-2. Write a program to seprate positive and negative integers form the list.

```python
L = [24,78,-64,-71,65,-87,-43,72,69,77,56,-90,-62]

L1 = []
L2 = []
for i in range(13):              # range me utni limit mention krte h jitne values vaiable m hoti hai.
    if L[i]>=0:
        L1.append(L[i])
    else:
        L2.append(L[i])
print("The list of positive number are",L1)
print("The list of negative number are",L2)
```

# Que-3. Write a program to seprate the even or odd number in the list.

```python
L = [24,53,87,62,56,90,45,32,78,57,82,61,54,43,28]

L1 = []
L2 = []
for i in range(15):
    if L[i]%2==0:
        L1.append(L[i])
    else:
        L2.append(L[i])

print("The list of even numbers is",L1)
print("The list of odd numbers is",L2)
```

# Que-1. Ask a user to enter a string and filter count vowels in the strings.Problem

```python
L = []
P = input("Enter the string : ")

for i in range (len(P)):
    L.append(P[i])
print(L)

count = 0

for i in range(len(L)):
    if L[i]=="a":
        print("a vowel is in the string.")
        count=count+1
        print(count)

for i in range(len(L)):
    if L[i]=="e":
        print("e vowel is in the string.")
        count=count+1
        print(count)

for i in range(len(L)):
    if L[i]=="i":
        print("i vowel is in the string.")
        count=count+1
        print(count)

for i in range(len(L)):
    if L[i]=="o":
        print("o vowel is in the string.")
        count=count+1
        print(count)

for i in range(len(L)):
    if L[i]=="u":
        print("u vowel is in the string.")
        count=count+1
        print(count)
```
