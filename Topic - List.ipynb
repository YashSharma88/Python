# Que-1. Take the input of the list from the user and then reverse the list.

```python
a = []
n = int(input("Enter the number of elements you want in the list : "))
for i in range(n):
    a.append(int(input("Enter elements : ")))
print(a)

a.reverse()
print(a)
```

# Que-2. Take the input from the user and interchange the first and the last element of the list.

```python
l = []
m = int(input("How many items you want in the list : "))
for i in range(m):
    l.append(int(input("Enter elements : ")))
print(l)

temp = l[0]
l[0] = l[4]
l[4] = temp
print(l)
```
