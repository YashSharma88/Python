# How to use variable-length arguments in a function in python?

```python
# Variable number of input arguments:-
```

```python
# Ex-

def argfunc(*args):
    s=0
    for i in range(len(args)):
        s=s+args[i]
    print(s)
```

```python
argfunc(14,32.5,56.3,78,23.4,67.32,65.34,22.45,28.32,12.38,76)
```

```python
# Ex-

def practical(**kwargs):
    for x in kwargs:
        print(kwargs, kwargs[x])
```

```python
practical(a='Lion', b=2.3, c=24)
```

# Generator Function :-

```python
# Ex- 

def test_fib(n):
    a=0
    b=1
    for i in range(n):
        yield a
        a,b=b , a+b
```

```python
test_fib(10)
```

```python
for i in test_fib(10):
    print(i)
```

# Lambda Function :-

```python
# Ex-

n=3
p=1

def test(n,p):
    return n**p

# 2nd Method :-

a=lambda n,p : n**p
```

```python
test(2,3)
```

```python
a(3,4)
```

```python
# Ex-

add=lambda x,y : x+y
```

```python
add(3,7)
```

```python
v=lambda c : (9/5)*c + 32

# Formula :- F = (9/5)*c + 32

v(45)
```

```python
k=lambda g,h : g//h  # (// --> Remainder)
```

```python
k(4,3)
```
