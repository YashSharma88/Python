# Dictionary :-

```python
# Ex - 
a = {"colour" : "red",
     "year" : 1997,
    "sports" : "cricket"}
print(a)
print(type(a))
```

```python
a = {"colour" : "red",
     "year" : 1997,
     "sports" : "cricket"}
print(a)
print(type(a))
print(a['year'])
```

# Duplicates not allowed :-

```python
a = {"colour" : "red",
     "year" : 1997,
     "sports" : "cricket",
     "sports" : "basketball",
     "sports" : "football"}        # Only last value consider in duplicates in dictionary.
print(a)   
```

# Dictionary length :-

```python
d = {"apple" : "fruits", "red" : "colour", "north" : "direction", 2:"code"}
print(len(d))
```

# Dictionary item - data types :-

```python
a = {"colour" : "red", 12 : 14, 2 : 56.3, "apple" : [1,2,3], 3 : (-1,-2,-3), "boy" : [2,4,6]}
print(a)
print(type(a))
```

# Dict() constructor :-

```python
n = dict(name = "Priya", age = 24, gender = "female")
print(n)
print(type(n))
```

# Accessing items :-

==> You can access the items of a dictionary by refering to its key name inside the square bracket.

```python
n = {1 : "apple", 2 : "boy", 3 : "cat"}
print(n)
print(n[1])
print(n[2])
print(n[3])
```

# get() method :-

```python
n = {1 : "apple", 2 : "boy", 3 : "cat"}
x = n.get(2)
print(x)
```

# keys() :-

```python
m = {"apple" : 12, 25 : "hello", 26.3 : "cat", "python" : -12}
x = m.keys()
print(x)
print(type(x))
```

# Values() :-

```python
m = {"apple" : 12, 25 : "hello", 26.3 : "cat", "python" : -12}
x = m.values()
print(x)
print(type(x))
print("\n")

# Or

print(m.values())
```

# items() :-

```python
m = {"apple" : 12, 25 : "hello", 26.3 : "cat", "python" : -12}
x = m.items()
print(x)
print(type(x))
```

# Check if key exist :-

```python
m = {"apple" : 12, 25 : "hello", 26.3 : "cat", "python" : -12}
if 25 in m:
    print("Yes")

if "boy" in m:
    print("Yes")
else:
    print("No")
```

# Change Value :-

```python
b = {10 : "Hello", 11 : "World", 9 : [12,13,14]}
b[9] = ["a","b","c"]
print(b)
```

# Removing() | pop() method :-

```python
a = {10 : "Create", 11 : "Hey", 12 : "Yash"}
a.pop(11)
print(a)
```

# popitem() method :-

```python
a = {10 : "create", 11 : "add", 12 : (1,2,3)}
a.popitem()             # By default last item has been remove in the dictionary.
print(a)
```

# del() method :-

```python
a = {"a" : "hello", "b" : "world", "c" : (10,11,12)}
del a["b"]
print(a)
```

```python
del a         # delete the dictionary of a.
print(a)
```

# clear() method :-

```python
a = {1 : "Himachal", 2 : "Jaipur", 3 : 35}
a.clear()
print(a,"\n",type(a))
```

# copy() method :-

Que- Why you need copy method?

```python
a = {1 : [23,25,28], "code" : (-1,-2,-3), "Dubai" : "Best country"}
b = a
print(b)
a.update({"India" : "Best country in the world."})
print(a)
print(b)
```

```python
a = {1 : [23,25,28], "code" : (-1,-2,-3), "Dubai" : "Best country"}
b = a.copy()
print(b)
a.update({"India" : "Best country in the world."})
print(a)
print(b)
```

# Make a copy of a dictionary with the dict() function :-

```python
a = {1 : "colours", 2 : "India", 3 : [10,11,12]}
b = dict(a)
print(b)
a.update({3 : "education"})
print(a)
print(b)
```

# Nested dictionaries :-

```python
a = { "class1" : {"Name" : "Ansul", "Section" : "A", "age" : 24},
      "class2" : {"Name" : "Yashu", "Section" : "B", "age" : 21},
      "class3" : {"Name" : "Virat", "Section" : "C", "age" : 23}}
print(a)
print(a["class3"]["Name"])
```

# fromkeys() method :-

```python
# Ex- 

x = {"Key1", "Key2", "Key3"}
y = 0
thisdict = dict.fromkeys(x,y)
print(thisdict)
```

```python
x = {"Key1", "Key2", "Key3"}
a = dict.fromkeys(x)
print(a)
```

# setdefault() method :-

```python
# Ex- 

movie = {"genre" : "comedy",
         "title" : "spiderman",
         "year"  : 2000}
x = movie.setdefault("title","xyz")
print(x)
```

```python
movie = {"genre" : "comedy",
         "title" : "spiderman",
         "year"  : 2000}
x = movie.setdefault("length","2 hours")
print(x)
print(movie)
```

```python
# Sol - 
a = {}
n = int(input("Enter the number of items you want in the dictionary : "))
for i in range(n):
    b = input("Enter key : ")
    c = input("Enter value : ")
    a.update({b:c})
print(a)

m = list(a.items())
print(m)
m.reverse()
print(m)
print(dict(m))
```
