# 2. lower() and upper() method :-

```python
# lower() method Ex :-

a = "How are you DOIng?"
b = a.lower()
print(b)
```

```python
a = "How are you DOIng?"
b = "Whats your plan for SUNDAy?"

print(a.lower() +" "+ b.lower())
```

```python
# upper() method Ex :-

a = "My name is yash, who are you?"
print(a.upper())
```

```python
b = "Practise makes a men perfect,"
c = "Yash you can do it."
print(b.upper() +" "+c.upper())
```

```python
x = "             Honesty is the best policy.   "
y = "      Honest man is always good.       "
z = "         We aPPRICIATE HonestY."

m = x.strip()
n = y.strip()
o = z.strip()
print(m.lower() +" "+ n.lower() +" "+ o.lower())
```
